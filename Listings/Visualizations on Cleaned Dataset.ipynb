## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv('cleaned_listings.csv')
```

## 2. Basic Overview

```python
df.info()
df.describe()
df.head()
```

## 3. Missing Values Heatmap

```python
sns.heatmap(df.isnull(), cbar=False, cmap="YlGnBu")
plt.title("Missing Data Heatmap")
plt.show()
```

## 4. Price Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=40, kde=True)
plt.title("Distribution of Listing Prices")
plt.xlabel("Price ($)")
plt.ylabel("Count")
plt.show()
```

## 5. Room Type Count

```python
sns.countplot(x='room_type', data=df)
plt.title("Room Types Distribution")
plt.xlabel("Room Type")
plt.ylabel("Number of Listings")
plt.xticks(rotation=45)
plt.show()
```

## 6. Average Price By Room Type

```python
plt.figure(figsize=(6,4))
sns.barplot(x='room_type', y='price', data=df)
plt.title("Average Price by Room Type")
plt.ylabel("Average Price ($)")
plt.show()
```

## 7. Availability Vs Price Scatterplot

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='availability_365', y='price', hue='room_type', data=df)
plt.title("Availability vs Price by Room Type")
plt.xlabel("Availability (days/year)")
plt.ylabel("Price ($)")
plt.show()
```

## 8. Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Numerical Features")
plt.show()
```

## 9. Top 10 Neigbourhoods by Listings

```python
top_neigh = df['neighbourhood'].value_counts().head(10)
top_neigh.plot(kind='barh', figsize=(8,5), color='skyblue')
plt.title("Top 10 Neighborhoods by Number of Listings")
plt.xlabel("Number of Listings")
plt.gca().invert_yaxis()
plt.show()
```
